```python
# prob1
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font',family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
my_random = np.random.normal(100,10,10000)
print(f'mean : {np.mean(my_random)}')
print(f'standard_deviance : {np.var(my_random)**(0.5)}')

my_random_mean=[np.mean(my_random[:i]) for i in range(1,len(my_random))]
my_random_var=[np.var(my_random[:i])**(0.5) for i in range(1,len(my_random))]

plt.plot(range(len(my_random_mean)),my_random_mean)
plt.title('  =>  100() ',fontsize=18)
plt.tight_layout()
plt.show()

plt.plot(range(len(my_random_var)),my_random_var)
plt.title('  => 10() ',fontsize=18)
plt.tight_layout()
plt.show()
```

```python
bi_10_05=np.random.binomial(10,0.5,10)
bi_30_05=np.random.binomial(30,0.5,30)
bi_50_05=np.random.binomial(50,0.5,50)
```

```python
plt.hist((bi_10_05 -5)/(2.5),alpha=0.3,label='Binomial_10_05',weights=np.zeros_like(bi_10_05)+1./bi_10_05.size)
plt.hist((bi_30_05 -15)/(7.5),alpha=0.3,label='Binomial_30_05',weights=np.zeros_like(bi_30_05)+1./bi_30_05.size)
plt.hist((bi_50_05 -25)/(12.5),alpha=0.3,label='Binomial_50_05',weights=np.zeros_like(bi_50_05)+1./bi_50_05.size)

mean = 0; std = 1; variance = np.square(std)
x = np.arange(-5,5,.01)
f = np.exp(-np.square(x-mean)/2*variance)/(np.sqrt(2*np.pi*variance))
plt.plot(x,f,lw=4,alpha=0.5,label='Normal_dist')

plt.tight_layout()
plt.legend()
plt.show()
```
