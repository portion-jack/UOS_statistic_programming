```python
# prob1
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font',family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
my_random = np.random.normal(100,10,10000)
print(f'mean : {np.mean(my_random)}')
print(f'standard_deviance : {np.var(my_random)**(0.5)}')

my_random_mean=[np.mean(my_random[:i]) for i in range(1,len(my_random))]
my_random_var=[np.var(my_random[:i])**(0.5) for i in range(1,len(my_random))]

plt.plot(range(len(my_random_mean)),my_random_mean)
plt.title('평균 그래프 =>  100(모평균)으로 수렴',fontsize=18)
plt.tight_layout()
plt.show()

plt.plot(range(len(my_random_var)),my_random_var)
plt.title('표준편차 그래프 => 10(모분산)으로 수렴',fontsize=18)
plt.tight_layout()
plt.show()
```

```python
from collections import Counter
bi_10_05=np.random.binomial(10,0.5,1000)
bi_30_05=np.random.binomial(30,0.5,1000)
bi_50_05=np.random.binomial(50,0.5,1000)

_bi_10_05=Counter(bi_10_05)
_bi_30_05=Counter(bi_30_05)
_bi_50_05=Counter(bi_50_05)
```

```python
for k,v in zip(dict(_bi_10_05).keys(),dict(_bi_10_05).values()):
    plt.bar((k-5)/(2.5)**(0.5),
            1.5*v/sum(_bi_10_05.values()),color='b',alpha=0.3)

for k,v in zip(dict(_bi_30_05).keys(),dict(_bi_30_05).values()):
    plt.bar((k-15)/(7.5)**(0.5),
            2.5*v/sum(_bi_30_05.values()),color='g',alpha=0.3)

for k,v in zip(dict(_bi_50_05).keys(),dict(_bi_50_05).values()):
    plt.bar((k-25)/(12.5)**(0.5),
            3.5*v/sum(_bi_50_05.values()),color='r',alpha=0.3)

mean = 0; std = 1; variance = np.square(std)
x = np.arange(-5,5,.01)
f = np.exp(-np.square(x-mean)/2*variance)/(np.sqrt(2*np.pi*variance))

plt.plot(x,f,lw=8,alpha=1,label='Normal_dist')

plt.tight_layout()
plt.legend()
plt.show()
```
